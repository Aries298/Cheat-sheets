```python
from bs4 import BeautifulSoup
import requests
```

```python
# Reading page content
url = '''https://en.wikipedia.org/wiki/Proto-Slavic_language'''
res = requests.get(url)
res.raise_for_status()
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.prettify())
```

```python
# Selecting element
link = soup.select('#mw-content-text > div.mw-parser-output > table.infobox > tbody > tr:nth-child(2) > td > a')
print(link)
```

```python
# Downloading images
img = '''https://upload.wikimedia.org/wikipedia/commons/thumb/a/a1/Balto-Slavic_lng.png/220px-Balto-Slavic_lng.png'''
res = requests.get(img)
imageFile = open("C:\\Users\\Admin\\Desktop\\testimg.png", 
'wb')
for chunk in res.iter_content(100000):
    imageFile.write(chunk)
imageFile.close()
```
